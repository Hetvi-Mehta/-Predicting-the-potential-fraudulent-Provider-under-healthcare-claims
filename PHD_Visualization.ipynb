```python
merge3['AdmissionDt'] = pd.to_datetime(merge3['AdmissionDt'] , format = '%Y-%m-%d')
merge3['DischargeDt'] = pd.to_datetime(merge3['DischargeDt'],format = '%Y-%m-%d')
merge3['ClaimStartDt'] = pd.to_datetime(merge3['ClaimStartDt'],format = '%Y-%m-%d')
merge3['ClaimEndDt'] = pd.to_datetime(merge3['ClaimEndDt'],format = '%Y-%m-%d')
merge3['DOB'] = pd.to_datetime(merge3['DOB'],format = '%Y-%m-%d')
merge3['DOD'] = pd.to_datetime(merge3['DOD'],format = '%Y-%m-%d')
```

### Correlation matrix

```python
import numpy as np
import seaborn as sns
# Generate a mask for the upper triangle
mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(8,8))
    
# Generate a custom diverging colormap
cmap = sns.diverging_palette(1000, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlations, mask=mask, cmap=cmap, vmax=.5, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
